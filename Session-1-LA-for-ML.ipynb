<h3> Some Matrix Operation </h3>

```python
# Matrix additon 

mat_1 = np.array([
    [1,2],
    [3,4]
])

mat_2 = np.array([
    [-1,-2],
    [-3,-4]
])

result_1 = mat_1 + mat_2

# Matrix subtraction

result_2 = mat_1 - mat_2

# Scalar multiplication

result_3 = 3*mat_1

# Matrix multiplication

result_4 = mat_1 @ mat_2

print(f'Matirx addition:\n {result_1}')

print(f'Matrix subtraction:\n {result_2}')

print(f'Matrix multiplication:\n {result_4}')

print(f'scalar multiplication:\n {result_3}')
```

```python
# Tensor product 

# They are useful to mathematically represent the composite state of multiple quantum state

arr_1 = np.array([
    [1,2],
    [3,4]
])

arr_2 = np.array([
    [4,5],
    [6,7]
])

np.kron(arr_1,arr_2) # for generating tensor product
```

```python
# Dot Product

vec_1 = np.array([1,2,3])

vec_2 = np.array([4,5,6])

np.dot(vec_1, vec_2) # for generating dot product of two vectors 
```

```python
# Inner product ( way - 1 )

# for inner product first vector should be bra(Transposed complex conjugate of column matrix)

vec_1_new = vec_1[np.newaxis,:].conj() # first convert to row matrix  then conjugate the values

vec_2_new = vec_2[:,np.newaxis] # creating column matrix from (3,) to (3,1)

inner_prod = vec_1 @ vec_2

print(inner_prod)
```

```python
# Inner product ( way - 2 )

np.inner(vec_1.conj(),vec_2) # using library function
```

```python
# outer product ( way-1 )


state_1 = np.array([1,2,3])

state_2 = np.array([4,5,6])

state_1_new = state_1[:,np.newaxis] # converting (3,) to (3,1)

state_2_new = state_2[np.newaxis,:].conj() # first row matrix (3,) to (1,3) then complex conjugate

outer_prod = state_1_new @ state_2_new

print(outer_prod)
```

```python
# outer product ( way-2 )

np.outer(state_1 , state_2.conj()) # using library function
```

<h3> Some Important Matices </h3>

```python
# Hermitian Matrix
# if A(dagger) = A

def check_hermitian(matrix):
    
    matrix_dagger = matrix.T.conj() # A(dagger) = (A^T)^*
    
    are_equal = np.allclose(matrix_dagger,matrix) # to check whether two matrix is same or not
    
    return are_equal

matrix_1 = np.array([
    [complex(5,0),complex(3,7)],
    [complex(3,-7),complex(2,0)]
])

matrix_2 = np.array([
    [1,2],
    [3,4]
])

print(f'Matrix_1 is hermitian ? = {check_hermitian(matrix_1)}')

print(f'Matix_2 is hermitian ? = {check_hermitian(matrix_2)}')
```

```python
# Unitary matrix
# If A(dagger) = A^-1 or => A(dagger)A = I(identity matrix)

def check_unitary(matrix):
    
    matrix_dagger = matrix.T.conj()
    
    identity_mat = np.eye(matrix.ndim) # or can use np.eye(len(matrix)) , ndim means dimension
    
    res = matrix_dagger @ matrix 
    
    are_equal = np.allclose(res,identity_mat)
    
    return are_equal

matrix_1 = np.array([
    [0,1],
    [1,0]
])

matrix_2 = np.array([
    [1,2],
    [3,4]
])

print(f'Matrix_1 is hermitian ? = {check_unitary(matrix_1)}')

print(f'Matix_2 is hermitian ? = {check_unitary(matrix_2)}')
```

<h5> Solving Linear Equation </h5>

A) **system of linear equations** (or **linear system**) is a collection of one or more linear equations involving the same variables. For example:


$\begin{cases}
-x_1+3x_2=7, \\ 3x_1+2x_2=1, \end{cases}\tag{1}$

is a system of two equations with two unknown variables $x_1$, $x_2$. **To solve** a system of linear equations means to find such values of the variables $x_1$, $x_2$, that all of its equations are simultaneously satisfied.

A linear system is **inconsistent** if it has no solution, and otherwise it is said to be **consistent**. Consistent system can have one or infinite number of solutions.

```python
A_system = np.array([
    [-1,3],
    [3,2]
])

b = np.array([7,1])

```

```python
# solving ( way-1 )

np.linalg.solve(A_system,b)
```

<b> So this is a consistent , complete and non-singular system </b>

<b><i> There is another method which is called ( Guassian Elimination Method ) which will be covered in next session </i></b>

B) Given another system of linear equations:

$\begin{cases}
-x_1+3x_2=7, \\ 3x_1-9x_2=1, \end{cases}\tag{5}$

let's find the determinant of the corresponding matrix.

```python
B_system = np.array([
    [-1,3],
    [3,-9]
])

b = np.array([7,1])
```

```python
# getting determinant
np.linalg.det(B_system)
```

```python
# let's see this system solution
try:
    np.linalg.solve(B_system,b)
except np.linalg.LinAlgError as err:
    print(err)
```

<b> This is because the sysem is contradictory and singular </b>

C) A system with infinite number of solutions
$(

```python
C_system = np.array([
    [-1,3],
    [3,-9]
])

b = np.array([7,21])
```
